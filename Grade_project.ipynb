```python
import gspread
import pandas as pd
```

```python
data = pd.read_csv("Grade Distribution 2012-2023 Queens College.xlsx - s2024.csv")
data.head()
```

```python
# Calculate average GPA by year
average_gpa_by_year = data.groupby('TERM')['AVG GPA'].mean()
print(average_gpa_by_year)
```

```python
# Calculate correlation between total enrollment and average GPA
correlation = data[['TOTAL', 'AVG GPA']].corr()
print(correlation)
```

```python
# Calculate average GPA by professor
average_gpa_by_prof = data.groupby('PROF')['AVG GPA'].mean()
print(average_gpa_by_prof)
```

```python
import matplotlib.pyplot as plt

# Adjust figure size for better visibility
plt.figure(figsize=(10, 6))

# Histogram of average GPAs
plt.hist(data['AVG GPA'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Average GPA', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Average GPAs', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for clarity
plt.tight_layout()  # Adjust layout to avoid cut-off labels
plt.show()

# Bar chart of course enrollments
plt.figure(figsize=(12, 8))
plt.bar(data['COURSE NAME'], data['TOTAL'], color='teal')
plt.xlabel('Course Name', fontsize=12)
plt.ylabel('Enrollment', fontsize=12)
plt.title('Course Enrollment', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate labels for clarity
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Scatter plot of total enrollment vs. average GPA
plt.figure(figsize=(10, 6))
plt.scatter(data['TOTAL'], data['AVG GPA'], color='purple', alpha=0.7, edgecolor='black')
plt.xlabel('Total Enrollment', fontsize=12)
plt.ylabel('Average GPA', fontsize=12)
plt.title('Total Enrollment vs. Average GPA', fontsize=14)
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print("Summary Statistics:")
print(data.describe())

# Distribution of Average GPAs
plt.figure(figsize=(10, 6))
plt.hist(data['AVG GPA'], bins=8, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(data['AVG GPA'].mean(), color='red', linestyle='--', label=f"Mean GPA: {data['AVG GPA'].mean():.2f}")
plt.axvline(data['AVG GPA'].median(), color='orange', linestyle='-', label=f"Median GPA: {data['AVG GPA'].median():.2f}")
plt.xlabel('Average GPA', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Average GPAs', fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Scatter plot of total enrollment vs. average GPA
plt.figure(figsize=(10, 6))
plt.scatter(data['TOTAL'], data['AVG GPA'], color='purple', alpha=0.7, edgecolor='black')
plt.xlabel('Total Enrollment', fontsize=12)
plt.ylabel('Average GPA', fontsize=12)
plt.title('Total Enrollment vs. Average GPA', fontsize=14)
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Correlation Insights
correlation = data['TOTAL'].corr(data['AVG GPA'])
print(f"Correlation between Total Enrollment and Average GPA: {correlation:.2f}")
```
